```python
red.plot(kind='box',subplots=True,layout=(4,4),sharex=False)
```

```python
for col in red.columns:
    Q1=red[col].quantile(0.25)
    Q3=red[col].quantile(0.75)
    IQR=Q3-Q1
    UF=Q3+(1.5*IQR)
    LF=Q1-(1.5*IQR)
    median=red[col].median()
    red[col]=np.where((red[col]<LF) | (red[col]>UF),median,red[col])
```

```python
red.plot(kind='box',subplots=True,layout=(4,4),sharex=False)

#red outliet check: pass
```

```python
white.plot(kind='box',subplots=True,layout=(4,4),sharex=False)
```

```python
for col in white.columns:
    Q1=white[col].quantile(0.25)
    Q3=white[col].quantile(0.75)
    IQR=Q3-Q1
    UF=Q3+(1.5*IQR)
    LF=Q1-(1.5*IQR)
    median=white[col].median()
    white[col]=np.where((white[col]<LF) | (white[col]>UF),median,white[col])
```

```python
white.plot(kind='box',subplots=True,layout=(4,4),sharex=False)

#white outliet check: pass
```
